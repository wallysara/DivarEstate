## **7**

**در طول زمان قیمت‌های اسمی افزایش پیدا می‌کنند اما این افزایش لزوما به معنی بالارفتن ارزش واقعی ملک نیست و می‌تواند ناشی از تورم باشد. به ازای میانگین مبلغ قیمت(price_value) در سال‌های ۱۴۰۰ تا ۱۴۰۳ قیمت حقیقی را محاسبه کنید و بررسی کنید ترند قیمت حقیقی چگونه است. **


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1) پیش‌پردازش سال و فیلتر قیمت‌های معتبر
df = train.copy()
df["created_at_month"] = pd.to_datetime(df["created_at_month"], errors="coerce")
df = df[df["price_value"].notna() & (df["price_value"] > 0)]

# نگاشت تقریبی سال میلادی → شمسی (۱۴۰۰ تا ۱۴۰۳)
gregorian_to_jalali = {2021:1400, 2022:1401, 2023:1402, 2024:1403}
df["year_g"] = df["created_at_month"].dt.year
df = df[df["year_g"].isin(gregorian_to_jalali.keys())]
df["year_j"] = df["year_g"].map(gregorian_to_jalali)

# 2) میانگین سالانه قیمت اسمی (۱۴۰۰ تا ۱۴۰۳)
nominal = df.groupby("year_j", as_index=True)["price_value"].mean().sort_index()

# 3) استخراج «تورم سالانه» از روی میانگین قیمت اسمی
inflation_rate = nominal.pct_change()

# 4) ساخت شاخص قیمتی نسبی (CPI) با مبنای ۱۴۰۰=100
cpi = (1 + inflation_rate).cumprod()
cpi.loc[cpi.index.min()] = 1.0
cpi = cpi * 100

# 5) محاسبه قیمت حقیقی (بر حسب سال مبنا ۱۴۰۰)
real = nominal / (cpi / 100)

# 6) جدول خلاصه
summary = pd.DataFrame({
    "NominalMean": nominal,
    "InflationRate": inflation_rate * 100,
    "CPI(1400=100)": cpi,
    "RealMean_1400Base": real
})

print(summary.round(3))

# 7) ترسیم ترند قیمت اسمی و حقیقی
plt.figure(figsize=(12,6))
plt.plot(nominal.index, nominal.values, marker="o", label="Nominal Mean Price")
plt.plot(real.index, real.values, marker="o", label="Real Mean Price (Base=1400)")
plt.title("Nominal vs Real Mean Price (1400–1403)")
plt.xlabel("Jalali Year")
plt.ylabel("Price (Toman)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### فرمول های استفاده شده در مسئله سوال 7

ستون NominalMean: میانگین قیمت اسمی (همون عدد خام آگهی‌ها).

ستون InflationRate: نرخ تغییر سالانه بر اساس همین میانگین‌ها.

سال ۱۴۰۱ نسبت به ۱۴۰۰ → حدود ۳۰٪ کاهش!

سال ۱۴۰۲ نسبت به ۱۴۰۱ → حدود ۱۸٪ کاهش.

سال ۱۴۰۳ نسبت به ۱۴۰۲ → جهش ۱۱۱٪ افزایش.

ستون CPI: شاخص قیمتی نسبی با مبنای ۱۴۰۰=100.

ستون RealMean_1400Base: میانگین قیمت حقیقی بر مبنای ۱۴۰۰ → همیشه ثابت (≈ 1.435e+10).

این یعنی تغییرات قیمت اسمی (بالا/پایین رفتنش) تنها بازتاب تورم داخلی همین بازار بوده، نه تغییر در ارزش واقعی.

# **آزمون فرض**

داشتن سند تجاری (یا هر نوع سند ملکی) در املاک به این معنی است که سند مالکیت معتبر، رسمی و قانونی برای ملک تجاری دارید. این سند نشان می‌دهد که شما صاحب قانونی ملک تجاری هستید و می‌توانید از حقوق مالکیت آن استفاده کنید. بررسی کنید که آیا داشتن سند تجاری(has_business_deed) بر میانگین قیمت فروش ملک تجاری تاثیر معناداری دارد؟


```python
import pandas as pd
from scipy.stats import ttest_ind

# فیلتر ملک‌های تجاری فروش
df = train[(train["cat2_slug"] == "commercial-sell") & (train["price_value"].notna())]

# تقسیم به دو گروه
with_deed = df[df["has_business_deed"] == True]["price_value"]
without_deed = df[df["has_business_deed"] == False]["price_value"]

# آزمون t-test
t_stat, p_val = ttest_ind(with_deed, without_deed, equal_var=False, nan_policy='omit')

print("t-statistic:", t_stat)
print("p-value:", p_val)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.boxplot(x="has_business_deed", y="price_value", data=df)
plt.yscale("log")
plt.xticks([0,1], [reshaper("بدون سند تجاری"), reshaper("با سند تجاری")])
plt.title(reshaper("مقایسه قیمت فروش ملک‌های تجاری بر اساس داشتن سند تجاری"))
plt.xlabel(reshaper("سند تجاری"))
plt.ylabel(reshaper("قیمت فروش (تومان)"))
plt.show()

```
